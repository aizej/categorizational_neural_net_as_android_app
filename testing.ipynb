```python
from tensorflow import keras
import cv2
import matplotlib.pyplot as plt
import numpy as np
import time
```

```python
model = keras.models.load_model(r'C:\Users\david\Desktop\kodezy\python\normalni_projekty\AI_2typesof images\model2_catsVSdogs_5epoch.h5')
```

```python
time.sleep(5)
cap = cv2.VideoCapture(1)

def read_from_VideoCapture(capture):
    ret, frame = cap.read()
    ret, frame = cap.read()
    ret, frame = cap.read()
    return ret, frame

ret, frame = read_from_VideoCapture(cap)

#cv2.imshow('frame',frame)
#cv2.waitKey()


cap.release()
cv2.destroyAllWindows()

CATEGORIES = ["Cat","Dog"]


my_image = frame
#my_image = cv2.imread(r"C:\Users\Uzivatel\Desktop\kodezy\python\normalni_projekty\AI_2typesof images\test\exotic1.jpg")
my_image = cv2.resize(my_image, (150, 150))
my_image = my_image/255

prediction = model.predict(np.array([my_image]))
prediction = prediction[0].tolist()
print(prediction)

print(CATEGORIES[prediction.index(max(prediction))],f"  {max(prediction) * 100}%\n")
plt.imshow(my_image)
```
